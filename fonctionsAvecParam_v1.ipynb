## 2 - Param√®tres

## ‚öôÔ∏è Param√®tres d'une fonction

Les **param√®tres** sont des **variables** utilis√©es par une fonction.  
Ils sont plac√©s **entre les parenth√®ses** lors de la d√©finition de la fonction.

---

### üìå Structure g√©n√©rale

~~~python
def nomFonction(parametre):
    bloc_d_instructions
~~~

---

### üß© Exemple simple

~~~python
def dire_bonjour(prenom):
    print("Bonjour", prenom)
~~~

Ici :
- `dire_bonjour` est le **nom de la fonction** ;
- `prenom` est le **param√®tre** ;
- la fonction utilise ce param√®tre dans ses instructions.

On appelle la fonction ainsi:

~~~python
dire_bonjour("Bob")
~~~
- la valeur "Bob" est affect√© au param√®tre `prenom` de la fonction `dire_bonjour`;
- le texte suivant est affich√© dans la console: Bpnjour Bob.

---

## üéØ But des param√®tres

Les **param√®tres** permettent d‚Äô**envoyer des donn√©es (des valeurs) depuis l‚Äôappel d‚Äôune fonction vers la d√©finition de la fonction**.

Lors de l‚Äôappel, une **affectation invisible** se fait :
- `param` re√ßoit la `valeur` en **argument** lors de l'appel de fonction.

La fonction devient ainsi **plus g√©n√©rale et r√©utilisable** : on parle de **g√©n√©ricit√©**.

---

## üìñ Vocabulaire

Lors de l'appel de la fonction, le param√®tre prend le nom d‚Äô**argument**.

## üîπ Fonction avec plusieurs param√®tres

Si une fonction poss√®de **plusieurs param√®tres**, on les s√©pare par une **virgule** dans la **d√©finition** **et** dans l‚Äô**appel** de la fonction.

### D√©finition de la fonction
~~~python
def nomFonction(param1, param2):
    bloc_d_instructions
~~~

### Appel de la fonction
~~~python
nomFonction(valeur1, valeur2)
~~~

Fonction d√©j√† connue avec plusieurs param√®tres:
print(param1, param2, param3, ....)

---

## üí° √Ä retenir

- Les **param√®tres sont des variables**
- Une **affectation invisible** se fait lors de l‚Äôappel de la fonction :
  - `param1` re√ßoit `valeur1`
  - `param2` re√ßoit `valeur2`
- Les param√®tres et les arguments sont **s√©par√©s** par des **virgules**
