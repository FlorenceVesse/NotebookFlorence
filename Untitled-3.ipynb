## üéØ But des param√®tres

Les **param√®tres** permettent d‚Äô**envoyer des donn√©es depuis l‚Äôappel d‚Äôune fonction vers la d√©finition de la fonction**.
Les **param√®tres** sont des **variables**.
- Une **affectation invisible** se fait lors de l‚Äôappel de la fonction :
  - `param` re√ßoit `valeur`

Les param√®tres peuvent √™tre utilis√©s, manipul√©s dans le code de la fonction.

La fonction va ainsi devenir plus g√©n√©ral, on parle de **g√©n√©ricit√©**.

---


## üîπ Fonction avec plusieurs param√®tres

Si une fonction poss√®de **plusieurs param√®tres**, on les s√©pare par une **virgule** dans la **d√©finition** **et** dans l‚Äô**appel** de la fonction.

### D√©finition de la fonction
~~~python
def nomFonction(param1, param2):
    bloc_d_instructions
~~~

### Appel de la fonction
~~~python
nomFonction(valeur1, valeur2)
~~~

---

## üí° √Ä retenir

- Les **param√®tres sont des variables**
- Une **affectation invisible** se fait lors de l‚Äôappel de la fonction :
  - `param1` re√ßoit `valeur1`
  - `param2` re√ßoit `valeur2`


### Exercice 603

Teste le code suivant et ajoute le code pour dessiner un rectangle suppl√©mentaire de longueur 100 et de hauteur 30.


````python
def rectangle(longueur, hauteur):
    for _ in range(2) :
        forward(longueur)
        left(90)
        forward(hauteur)
        left(90)

rectangle(30, 100)

done()
````
