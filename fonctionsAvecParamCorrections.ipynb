## 2 - Paramètres

## ⚙️ Paramètres d'une fonction

Les **paramètres** sont des **variables** utilisées par une fonction.  
Ils sont placés **entre les parenthèses** lors de la définition de la fonction.

---

### 📌 Structure générale

~~~python
def nomFonction(parametre):
    bloc_d_instructions
~~~

---

### 🧩 Exemple simple

~~~python
def dire_bonjour(prenom):
    print("Bonjour", prenom)
~~~

Ici :
- `dire_bonjour` est le **nom de la fonction** ;
- `prenom` est le **paramètre** ;
- la fonction utilise ce paramètre dans ses instructions.

---

## 🎯 But des paramètres

Les **paramètres** permettent d’**envoyer des données depuis l’appel d’une fonction vers la définition de la fonction**.

Lors de l’appel, une **affectation invisible** se fait :
- `param` reçoit `valeur`

La fonction devient ainsi **plus générale et réutilisable** : on parle de **généricité**.

---

## 📖 Vocabulaire

Lors de l'appel de la fonction, le paramètre prend le nom d’**argument**.

## 🔹 Fonction avec plusieurs paramètres

Si une fonction possède **plusieurs paramètres**, on les sépare par une **virgule** dans la **définition** **et** dans l’**appel** de la fonction.

### Définition de la fonction
~~~python
def nomFonction(param1, param2):
    bloc_d_instructions
~~~

### Appel de la fonction
~~~python
nomFonction(valeur1, valeur2)
~~~

Fonction déjà connue avec plusieurs paramètres:
print(param1, param2, param3, ....)

---

## 💡 À retenir

- Les **paramètres sont des variables**
- Une **affectation invisible** se fait lors de l’appel de la fonction :
  - `param1` reçoit `valeur1`
  - `param2` reçoit `valeur2`
- Les paramètres et les arguments sont **séparés** par des **virgules**

### 🔁 Exercice  11 — Polygone régulier

Complète la fonction pour tracer un polygone régulier.

Rappel :  
angle = 360 / nombre_de_côtés

Utilise cette fonction pour tracer : 

    - un carré de côté 30
    - un triangle de côté 50
    - un hexagone de côté 100


````python
from turtle import *


def polygone(nb_cotes, taille_cote):
    for i in range(nb_cotes):
        forward(taille_cote)
        left(360/nb_cotes)

# reculer la tortue pour voir le dessin        
up()
fd(-100)
down()

polygone(5, 80)   # pentagone
polygone(4, 30)   # carré
polygone(3, 50)   # triangle
polygone(6, 100)   # hexagone


done()
````
